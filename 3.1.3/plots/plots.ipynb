```python
import numpy as np
import tikzplotlib as tikz
import matplotlib.pyplot as plt
import scipy.optimize as sp
```

```python
T = np.array([3.21, 2.89, 2.55, 2.38, 1.90, 2.13, 0.87, 1.60, 1.4, 1.16])
N = np.array([12, 11, 10, 9, 7, 8, 3, 6, 5, 4])
T_err = 0.1
```

```python
def fun(x, p):
    return x*p
```

```python
k, err = sp.curve_fit(fun, N, T)
x = np.linspace(2, 13, 2)
plt.plot(x, fun(x, k), 'black')
plt.errorbar(N, T, T_err, fmt='.', color='red')
print(k, err)
#tikz.save('plot1.tex')
```
