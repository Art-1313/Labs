```python
import numpy as np
import scipy as sp
import scipy.optimize as fit
import matplotlib.pyplot as plt
import tikzplotlib as tikz
```

```python
def print_err(a, a_err):
    print(np.round(a, decimals = 1 + round(abs(np.log10(a_err)))),
    '\pm', np.round(a_err, decimals = 1 + round(abs(np.log10(a_err)))))
```

Первый график

```python
T_e = np.array([0.4, 0.6, 0.86, 1.25, 1.4, 1.6, 1.9, 2.0])
T_e_err = np.array([0.2, 0.2, 0.1, 0.3, 0.5, 0.5, 0.5, 0.6])
T_t = np.array([0.4, 0.7, 0.9, 1.2, 1.3, 1.6, 1.9, 2.0])
T_t_err = 0.1 * np.array([0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02])
```

```python
def func(x, p0):
    return p0 * x
```

```python
k, err = fit.curve_fit(func, T_e, T_t)
x = np.linspace(0, T_e[T_t.size - 1], 2)
plt.plot(x, func(x, k[0]), 'black')
plt.errorbar(T_e, T_t, T_t_err, T_e_err, fmt = '.', color = 'red')
#tikz.save("tex/fer_plot.tex")
for i in range(k.size):
    print_err(k[i], err[i][i])
```

Второй график

```python
Theta = np.array([8.78, 6.40, 4.38, 3.83, 2.61, 2.27])
Theta_err = np.array([0.70,	0.57, 0.47,	0.45, 0.37, 0.36])
R = np.array([0.64, 0.44, 0.33, 0.25, 0.19, 0.16])
```

```python
def func(x, p0):
    return p0 * x
```

```python
k, err = fit.curve_fit(func, R, Theta)
x = np.linspace(0, R[0], 2)
plt.plot(x, func(x, k[0]), 'black')
plt.errorbar(R, Theta, Theta_err, fmt = '.', color = 'red')
tikz.save("tex/plot2.tex")
for i in range(k.size):
    print_err(k[i], err[i][i])
```
